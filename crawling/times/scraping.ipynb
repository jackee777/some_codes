```python
car_name = []
urls = []
for i, j in enumerate(soup.findAll("a", { "class" : "alpha" })):
    if i <= 1:
        continue
    try:
        if j.findAll("img")[0].get("alt") == "駐車場経営による土地活用のことならタイムズ２４":
            break
        elif j.findAll("img")[0].get("alt") == "":
            car_name.append("NV350キャラバン（NISSAN)")
        else:
            car_name.append(j.findAll("img")[0].get("alt"))
        continue
    except:
        pass
    if j["href"][:7] == "/view/s":
        #print(j["href"])
        urls.append("https://plus.timescar.jp/"+j["href"][1:])
    
#car_name
urls
```

```python
all_arr = []
for i in tqdm(urls):
    time.sleep(10)
    count = 0
    car_arr = []
    r = requests.get(i)
    r.encoding = "utf-8"
    soup = BeautifulSoup(r.text, "html.parser")
    while True:
        for j in soup.findAll("span", { "id" : "adr1" }):
            car_arr.append(j.text)
        try:
            if soup.findAll("span", { "id" : "isDispNext" })[0].text == "次へ":
                time.sleep(10)
                r = requests.get(i.replace("list.jsp?",soup.findAll("span", { "id" : "isDispNext" })[0].next["href"]+"&"))
                r.encoding = "utf-8"
                #print(i.replace("list.jsp?",soup.findAll("span", { "id" : "isDispNext" })[0].next["href"]+"&"))
                soup = BeautifulSoup(r.text, "html.parser")
            else:
                break
        except IndexError:
            break
    all_arr.append(car_arr)
```
