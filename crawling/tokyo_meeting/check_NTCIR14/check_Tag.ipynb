```python
from collections import defaultdict
import json
```

# training 用の json

```python
train = json.load(open("Segmentation-FormalRun-Training.json", "r", encoding="utf8"))
train[:5]
```

# 行ごとに分割された会議録

```python
train_text = json.load(open("Pref13_tokyo.json", "r", encoding="utf8"))
train_text[:5]
```

# Question

### start

```python
Phrase_count = defaultdict(lambda : 0)
Feature_count = defaultdict(lambda : 0)
for js in train:
    line = js["QuestionStartingLine"]
    text_js = train_text[line-1]
    if line == text_js["Line"]:
        if text_js["Title"] == "平成24年第１回定例会(第２号)":
            print(text_js["Title"])
            # print("前", train_text[line-2]["Utterance"])
            print(line, text_js["Utterance"])
            # print("後", train_text[line]["Utterance"])
            print()
        Phrase_count[text_js["Utterance"]] += 1
        Feature_count[text_js["Utterance"].split("、")[0]] += 1
```

```python
sorted(Phrase_count.items(), key=lambda kv: kv[1], reverse=True)
```

```python
sorted(Feature_count.items(), key=lambda kv: kv[1], reverse=True)
```
