```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
np.random.seed(0)
```

```python
plt.figure(figsize = (16, 9))
mickey = np.random.normal(10, 5, 10000)
king = np.random.normal(20, 5, 10000)
plt.hist(mickey, bins=100, alpha = 0.5)
plt.hist(king, bins=100, alpha = 0.5)
plt.hist(np.append(mickey, king), bins=100, histtype="step", color = "red", linewidth=2)
plt.ylabel("cooccurence", fontsize=24)
plt.xlabel("cooccurence word", fontsize=24)
```
